## ISP Maturity - ASN ranking from caida, RDAP registry date, cone sizes, Prefix age from dat file ? 

```python
import pandas as pd
import requests
```

```python
test_df = pd.read_csv("remedies.csv", sep=';')
count_remedy = test_df.value_counts("ASN").reset_index()
count_remedy.columns=["ASN", "count"]
data = count_remedy
asn_details = {}
f=0
```

```python
data['ASN'] = count_remedy.apply(lambda x: int(x['ASN'].split()[0]),axis=1)
```

```python
def get_asn_metrics(asn : int)->{}:
    details = requests.get("https://api.asrank.caida.org/v2/restful/asns/" + str(asn))
    asn_details[asn] = details.json()
```

```python
for i in data['ASN']:
    get_asn_metrics(i)
    f=f+1
    print(f)
```

```python
def get_rank(asn):
    try:
        return asn_details[asn]['data']['asn']['rank']
    except TypeError:
        return 0
    
def get_asns_num(asn):
    try:
        return asn_details[asn]['data']['asn']['cone']['numberAsns']
    except TypeError:
        return 0
    
def get_prefixes_num(asn):
    try:
        return asn_details[asn]['data']['asn']['cone']['numberPrefixes']
    except TypeError:
        return 0
    
def get_cust_num(asn):
    try:
        return asn_details[asn]['data']['asn']['asnDegree']['customer']
    except TypeError:
        return 0
    
    
def get_addresses_num(asn):
    try:
        return asn_details[asn]['data']['asn']['cone']['numberAddresses']
    except TypeError:
        return 0
```

```python
data['cones_asns_number'] = data.ASN.apply(lambda x: get_asns_num(x))
data['rank'] = data.ASN.apply(lambda x: get_rank(x))
data['cones_prefixes_number'] = data.ASN.apply(lambda x: get_prefixes_num(x))
data['customer_number'] = data.ASN.apply(lambda x: get_cust_num(x))
data['addresses_number'] = data.ASN.apply(lambda x: get_addresses_num(x))
```

```python
rdap_metrics = {}
def get_rdap_metrics(asn):
    details = requests.get("https://rdap.arin.net/registry/autnum/" + str(asn)).json()
    rdap_metrics[asn]= details
```

```python
for i in data['ASN']:
    get_rdap_metrics(i)
    f=f+1
    print(f)
```

```python
recent_df = pd.read_csv("recent_tests.csv", sep=';', dtype={'Sesssion': 'str'})

#Resetting types
recent_df.astype({'Timestamp': 'str',
                'Client IP Block': 'str',
                'ASN': 'str', 
                'Country': 'str', 
                'NAT': 'str', 
                'Outbound Private Status': 'str',
               'Adjacent Spoof Prefix Length': 'str'}).dtypes

df_blocked = recent_df.loc[recent_df['Outbound Routable Status'] == 'blocked']
df_received = recent_df.loc[recent_df['Outbound Routable Status'] == 'received']
```

```python
spoofable_df = df_received.groupby('ASN')
spoofable_df = spoofable_df.count()[['Client IP Block']]
spoofable_df = spoofable_df.rename(columns={"Client IP Block" : "Spoofable Blocks"})

remedies_df = test_df.groupby('ASN')
remedies_df = remedies_df.count()[['IP Block']]
remedies_df = remedies_df.rename(columns={"IP Block" : "Blocks Fixed"})
final_df = spoofable_df.merge(remedies_df, on="ASN")
final_df = final_df.rename(columns={"IP Block": "Fixed Blocks"})
final_df['total'] = final_df['Spoofable Blocks'] + final_df['Blocks Fixed']
final_df['remediation rate'] = final_df['Blocks Fixed']/final_df['total']  # REMEDIATION RATE ACCORDING TO FORMULA
final_df = final_df.reset_index()
final_df['cones_asns_number'] = data.ASN.apply(lambda x: get_asns_num(x))
final_df['rank'] = data.ASN.apply(lambda x: get_rank(x))
final_df['cones_prefixes_number'] = data.ASN.apply(lambda x: get_prefixes_num(x))
final_df['customer_number'] = data.ASN.apply(lambda x: get_cust_num(x))
final_df['addresses_number'] = data.ASN.apply(lambda x: get_addresses_num(x))
final_df
```

```python
non_zero_cust_num = final_df[final_df['customer_number']!=0]
print("Remediation rate correlated with")
print("Number of prefixes in cone: ",non_zero_cust_num['remediation rate'].corr(non_zero_cust_num['cones_prefixes_number']))
print("Rank: ",final_df['remediation rate'].corr(final_df['rank']))
print("Number of ASNs in cone: ",final_df['remediation rate'].corr(final_df['cones_asns_number']))
print("Number of customers: ",final_df['remediation rate'].corr(final_df['customer_number']))
print("Number of addresses in cone: ",final_df['remediation rate'].corr(final_df['addresses_number']))
```
